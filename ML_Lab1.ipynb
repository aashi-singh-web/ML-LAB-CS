```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(arr.shape)
print(arr.dtype)
```

```python
import numpy as np

zeros_matrix = np.zeros((3, 4))
ones_matrix = np.ones((2, 5))

print(zeros_matrix)
print(zeros_matrix.shape)

print(ones_matrix)
print(ones_matrix.shape)
```

```python
import numpy as np

arr = np.arange(1, 13)
reshaped_arr = arr.reshape(3, 4)

print(arr)
print(arr.shape)

print(reshaped_arr)
print(reshaped_arr.shape)
```

```python
import numpy as np

arr = np.random.randint(1, 100, 10)

print(arr)
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.sum(arr))
```
